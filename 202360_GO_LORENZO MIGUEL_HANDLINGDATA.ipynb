```python
legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}

legs2 = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","adu"):{
         "travel_time_mins":55
     },
    ("adu","sbca"):{
         "travel_time_mins":15
     },
    ("sbca","upm"):{
         "travel_time_mins":5
     },
    ("upm","upd"):{
         "travel_time_mins":20
     }
}
def eta(legs,source,destination):
    total_travel_time = 0
    if source == destination:
        return total_travel_time
    elif tuple((source,destination)) in legs:
        total_travel_time += legs[tuple((source,destination))]["travel_time_mins"]
        return total_travel_time
    else:
        for i in legs:
            if i[0] == source:
                total_travel_time += legs[i]["travel_time_mins"]
            elif total_travel_time > 0:
                total_travel_time += legs[i]["travel_time_mins"]
                if i[1] == destination:
                    return total_travel_time
        for i in legs:
            if i[1] == destination:
                total_travel_time += legs[i]["travel_time_mins"]
                break
            total_travel_time += legs[i]["travel_time_mins"]
        return total_travel_time
                
        
eta(legs2,"dlsu","admu")
```
